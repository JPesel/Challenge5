# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "city_data.csv"
ride_data_to_load = "ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
type_ride_count = pyber_data_df.groupby(["type"]).count()["ride_id"].rename("Total Rides")
type_ride_count
```

```python
# 2. Get the total drivers for each city type
type_driver_sum = city_data_df.groupby(["type"]).sum()["driver_count"].rename("Total Drivers")
type_driver_sum.head()
```

```python
#  3. Get the total amount of fares for each city type
type_fare_sum = pyber_data_df.groupby(["type"]).sum()["fare"].rename("Total Fares")
type_fare_sum.head()
```

```python
#  4. Get the average fare per ride for each city type. 
avg_fare_ride = pd.Series(type_fare_sum/type_ride_count, name = "Average Fare per Ride")
avg_fare_ride
```

```python
# 5. Get the average fare per driver for each city type. 
avg_fare_driver = pd.Series(type_fare_sum/type_driver_sum, name = "Average Fare per Driver")
avg_fare_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.merge(type_ride_count, type_driver_sum, left_index=True, right_on='type', how="left")
pyber_summary_df = pd.merge(pyber_summary_df, type_fare_sum, left_index=True, right_on='type', how="left")
pyber_summary_df = pd.merge(pyber_summary_df, avg_fare_ride, left_index=True, right_on='type', how="left")
pyber_summary_df = pd.merge(pyber_summary_df, avg_fare_driver, left_index=True, right_on='type', how="left")

pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
#  8. Format the columns.
fpyber_summary_df = pd.DataFrame(pyber_summary_df['Total Rides'])
fpyber_summary_df['Total Rides'] = [f'{x:,}' for x in pyber_summary_df["Total Rides"]]
fpyber_summary_df['Total Drivers'] = [f'{x:,}' for x in pyber_summary_df["Total Drivers"]]
fpyber_summary_df['Total Fares'] = [f'${x:,.2f}' for x in pyber_summary_df["Total Fares"]]
fpyber_summary_df['Average Fare per Ride'] = [f'${x:,.2f}' for x in pyber_summary_df["Average Fare per Ride"]]
fpyber_summary_df['Average Fare per Driver'] = [f'${x:,.2f}' for x in pyber_summary_df["Average Fare per Driver"]]

fpyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
df = pyber_data_df.groupby(["type", "date"]).sum()[["fare"]]
df.head()
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
df.head()
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
df = df.reset_index()
df.head(5)
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
df_pivot = df.pivot(index="date", columns="type", values="fare")
df_pivot
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.

df_pivot_r = df_pivot.loc['2019-01-01':'2019-04-29']
df_pivot_r.tail(10)
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
df_pivot_r.index = pd.to_datetime(df_pivot_r.index)
df_pivot_r
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
df_pivot_r.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
resample_df = df_pivot_r.resample("W").sum()
resample_df.head(10)
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')


# Plot the data.
rural_df = resample_df["Rural"]
suburban_df = resample_df["Suburban"]
urban_df = resample_df["Urban"]


plt.subplots(figsize=(15, 6))

rural_df.plot(x="date", y="fare")
suburban_df.plot(x="date", y="fare")
urban_df.plot(x="date", y="fare")

plt.ylabel("Fare ($)")
plt.xlabel("")
plt.title("Total Fare by City Type")
plt.legend()

plt.show()

plt.savefig("analysis/Fig1.png")
```
